## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the line_subtotal for each line using the unit_price and the qty   
df['line_subtotal'] = df['unit_price'] * df['qty']   
df[['unit_price', 'qty', 'line_subtotal']].head(2)
```

```python
# Create a column for total weight.
df['total_weight'] = df['unit_weight'] * df['qty']

# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping_price'] = df['total_weight'].apply(lambda x: 7 * x if x > 50 else 10 * x)  

# Display the first three rows to verify
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = df['line_subtotal'] + df['shipping_price'] + (df['line_subtotal'] * 0.0925)

# Display the first three rows to verify
df[['line_subtotal', 'shipping_price', 'total_price']].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['cost_of_goods'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head(3)
```

```python
# Create a column for the profit of each line using line price and cost of goods
df['line_profit'] = df['line_price'] - df['cost_of_goods']
df.head(3)
```

## Part 3: Confirm your work
You have email receipts showing that the total prices for 3 orders. Confirm that your calculations match the receipts. Remember, each order has multiple lines.

Order ID 2742071 had a total price of \$152,811.89

Order ID 2173913 had a total price of \$162,388.71

Order ID 6128929 had a total price of \$923,441.25


```python
# Define the order IDs and their expected totals from email receipts
order_ids = [2742071, 2173913, 6128929]
expected_totals = {
    2742071: 152811.89,
    2173913: 162388.71,
    6128929: 923441.25
}

# Ensure the total_price column is numeric and handle missing values
df['total_price'] = pd.to_numeric(df['total_price'], errors='coerce')

# Check for missing values in the total_price column for the specified order IDs
missing_values = df[df['order_id'].isin(order_ids) & df['total_price'].isnull()]
if not missing_values.empty:
    print("Missing values found in total_price for the following rows:")
    print(missing_values)

# Filter data for the specified order IDs and sum the total price for each order
order_totals = df[df['order_id'].isin(order_ids)].groupby('order_id')['total_price'].sum()

# Print the total prices for each order and compare with expected totals
print("Order Totals:")
for order_id, total in order_totals.items():    
    print(f"Order ID {order_id}: ${total:,.2f}")
    if order_id in expected_totals:
        expected_total = expected_totals[order_id]
        if abs(total - expected_total) < 1e-2:  # Allowing a small tolerance for floating point comparison
            print(f"Order ID {order_id} matches the expected total.")
    else:
        print(f"Order ID {order_id} not found in the expected totals.")
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# Define the top 5 client IDs
top_5_client_ids = [33615, 66037, 46820, 38378, 24741]

# Check if the total_price column has any missing or incorrect values
print("Checking for missing values in total_price column:")
print(df['total_price'].isnull().sum())

# Ensure the total_price column is numeric and handle missing values
df['total_price'] = pd.to_numeric(df['total_price'], errors='coerce')

# Optionally, drop rows with NaN values in total_price after conversion
df = df.dropna(subset=['total_price'])

# Calculate the total revenue for the top 5 clients
top_5_revenue = df[df['client_id'].isin(top_5_client_ids)].groupby('client_id')['total_price'].sum()

# Print the total revenue for the top 5 clients
print("Total revenue for the top 5 clients:")
for client_id, revenue in top_5_revenue.items():
    print(f"Client ID {client_id}: ${revenue:,.2f}")

# Ensure the calculated revenues match the expected values
expected_revenues = {
    33615: 8377308.52,
    66037: 10259514.79,
    46820: 9743794.36,
    38378: 12906550.87,
    24741: 82268892.02
}

for client_id in top_5_client_ids:
    if client_id in top_5_revenue:
        calculated_revenue = top_5_revenue[client_id]
        expected_revenue = expected_revenues[client_id]
        if abs(calculated_revenue - expected_revenue) < 1e-2:  # Allowing a small tolerance for floating point comparison
            print(f"Client ID {client_id} matches the expected revenue.")
    else:
        print(f"Client ID {client_id} not found in the data.")

```

```python
# Create a summary DataFrame showing the totals for the top 5 clients with the following information:
# client_id, qty, shipping_price, line_price, line_cost, line_profit

# Define the top 5 client IDs
top_5_client_ids = [33615, 66037, 46820, 38378, 24741]

# Create a summary DataFrame for the top 5 clients
summary_df = df[df['client_id'].isin(top_5_client_ids)].groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'line_price': 'sum',
    'cost_of_goods': 'sum',
    'line_profit': 'sum'
}).rename(columns={
    'qty': 'qty',
    'shipping_price': 'shipping_price',
    'line_price': 'line_price',
    'cost_of_goods': 'line_cost',
    'line_profit': 'line_profit'
}).reset_index()

# Format the values to 2 decimal places
summary_df['shipping_price'] = summary_df['shipping_price'].map('{:.2f}'.format)
summary_df['line_price'] = summary_df['line_price'].map('{:.2f}'.format)
summary_df['line_cost'] = summary_df['line_cost'].map('{:.2f}'.format)
summary_df['line_profit'] = summary_df['line_profit'].map('{:.2f}'.format)

# Display the summary DataFrame
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns.
money_columns = ['shipping_price', 'total_price', 'cost_of_goods', 'line_profit']

# Define a function that converts a dollar amount to millions with 2 decimal places.
def currency_format_millions(x):
    return f"{x / 1_000_000:.6f}"

# Create a summary DataFrame showing the totals for the top 5 clients with the following information:
# client_id, qty, shipping_price, total_price, cost_of_goods, line_profit

# Filter the data for the top 5 clients
top_5_client_ids = [33615, 66037, 46820, 38378, 24741]
top_5_clients_data = df[df['client_id'].isin(top_5_client_ids)]

# Group by client_id and aggregate the required columns
summary_data = top_5_clients_data.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'cost_of_goods': 'sum',
    'line_profit': 'sum'
}).reset_index()

# Apply the currency_format_millions function to only the money columns.
summary_data[money_columns] = summary_data[money_columns].applymap(currency_format_millions)

# Rename the columns to reflect the change in the money format.
summary_data.columns = ['Client ID', 'Units', 'Shipping (millions)', 'Total Revenue (millions)', 'Total Cost (millions)', 'Total Profit (millions)']

# Display the updated DataFrame
summary_data
```

```python
# Sort the updated data by "Total Profit (millions)" from highest to lowest and assign the sort to a new DataFrame.
sorted_summary_data = summary_data.sort_values(by="Total Profit (millions)", ascending=False)

# Print the sorted DataFrame
print(sorted_summary_data)
```
